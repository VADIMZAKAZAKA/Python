```python
mat = [[1, 3, 5], [7, 9, 11], [13, 15, 17]]

for j in range(len(mat[0])):
    if all(mat[i][j] % 2 == 1 for i in range(len(mat))):
        print(j + 1)
        break
else:
    print(0)
```

```python
mat = [[3, 1, 4], [9, 2, 7], [5, 8, 6]]

for row in mat:
    a = row.index(min(row))
    b = row.index(max(row))
    row[a], row[b] = row[b], row[a]
    print(*row)
```
