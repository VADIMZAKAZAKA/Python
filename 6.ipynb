```python
lst = list(map(int, input().split()))
K = int(input())
L = int(input())

s = 0
count = 0
for i in range(len(lst)):
    if i < K-1 or i > L-1:
        s += lst[i]
        count += 1
print(s / count)
```

```python
lst = list(map(int, input().split()))
max_idx = lst.index(max(lst))
product = 1
for i in range(max_idx + 1, len(lst)):
    product *= lst[i]
print(product)
```

```python
lst = list(map(int, input().split()))
min_idx = lst.index(min(lst))
max_idx = lst.index(max(lst))
start = min(min_idx, max_idx)
end = max(min_idx, max_idx)
total = sum(lst[start:end+1])
print(total)
```

```python
lst = list(map(int, input().split()))
min_idx = lst.index(min(lst))
max_idx = lst.index(max(lst))
start = min(min_idx, max_idx)
end = max(min_idx, max_idx)
for i in range(start + 1, end):
    lst[i] = 0
print(*lst)
```

```python
lst = list(map(int, input().split()))
count = 0
for i in range(len(lst) - 1):
    if lst[i] > lst[i + 1]:
        print(i, end=' ')
        count += 1
print(count)
```

```python
lst = list(map(int, input().split()))
max_sum = lst[0] + lst[1]
idx = 0
for i in range(len(lst) - 1):
    if lst[i] + lst[i + 1] > max_sum:
        max_sum = lst[i] + lst[i + 1]
        idx = i
print(lst[idx], lst[idx + 1])
```

```python
lst = list(map(int, input().split()))
min_idx = lst.index(min(lst))
max_idx = lst.index(max(lst))
lst[min_idx], lst[max_idx] = lst[max_idx], lst[min_idx]
print(*lst)
```
